데이터 수집을 할 때 웹 개발 언어를 알아야 하는 이유는 웹 페이지에서 데이터를 가져오기 위해서입니다.

예를 들어, 어느 날 친구가 당신에게 "너는 컴퓨터가 얼마나 똑똑한지 알아?"라고 물어봤다고 생각해보세요. 당신은 궁금해졌고, 컴퓨터가 얼마나 똑똑한지 알아보기로 했습니다. 그래서 당신은 인터넷 검색을 통해 컴퓨터의 지능에 대한 정보를 찾아보기로 했습니다.

하지만 인터넷에는 수많은 정보가 있습니다. 모든 정보를 일일이 읽어보기에는 시간이 너무 오래 걸리죠. 이때 웹 개발 언어를 이해하고 파이썬 라이브러리를 사용하면 웹 페이지에서 필요한 정보만을 자동으로 추출할 수 있습니다.

## HTML, CSS, Javascript는 웹 개발에 사용되는 세 가지 주요 언어입니다.

- HTML은 HyperText Markup Language의 약자로 웹 페이지의 구조를 만들기 위한 언어입니다. <span style="background-color:#c6dbef">HTML은 웹 페이지의 내용과 구조를 정의하는 데 사용</span>됩니다.

- CSS는 Cascading Style Sheets의 약자로 웹 페이지의 디자인과 레이아웃을 만들기 위한 스타일시트입니다. <span style="background-color:#c6dbef">CSS는 HTML로 작성된 웹 페이지의 외형을 예쁘게 바꿀 수 있습니다</span>. 글자 색상과 배경 색상, 배치 등을 자유롭게 변경할 수 있습니다.

- Javascript는 웹 페이지를 프로그래밍하기 위한 언어입니다.<span style="background-color:#c6dbef"> Javascript를 사용하면 웹 페이지에 동적인 기능을 추가>/span>할 수 있습니다.

#### HTML의 구조  및 태그
- 구조 [f12누르면 보이는 정보들]
 - $<!Doctype html>$ : HTML5 문서를 선언하는 구문
 - $<html></html>$ : HTML 문서의 시작과 끝
 - $<head></head>$ : CSS(디자인이나 꾸미는거), JavaScript(동적인 처리), meta(데이터를 위한 정보), title(ex: 새탭이라고 뜨는것) 정보들을 설정
 - $<body></body>$ : 실제 홈페이지 화면을 나타내는 부분

- 요소 구조
 - HTML 요소는 여러 속성들을 가질 수 있으며 속성들은 해당 요소에 대한 추가 정보를 제공. (속성:테그에 달리는 안에 들어있는 속성)
 - 시작 태그\(< >)로 시작해서 종료 태그(</>)로 끝남
 - 요소 안에 다른 요소를 작성할 수 있음

```python
# HTML 기본 구조
html ='''
<!DOCTYPE html>                html5 명시
<html lang="en">               html 태그로 모든 태그를 감사줌. lang이라는 속성을 포함
 <head>                   head 태그는 meta, title 이외의 style, script, link와 같은 태그 포함
     <meta charset="UTF-8">    
     <title>a.html</title>      title 태그는 문서의 제목
     <meta name="viewport" content="width=device-width, initial-scale=1.0"> meta 태그는 name과 contents속성으로 문서 정보 포함
 </head> 
 <body>                        웹페이지의 내용 포함
     <style type="text/css">    CSS 코드가 포함된는 태그  
     </style>
                                script 태그는 JavaScript 코드를 작성하거나 파일 로드. head 태그 or body 태그 하단 위치
 </body>
</html>
'''
print(f'HTML 구조 \n {html}')
```

"viewport": 휴대폰으로 봤을 때 사이즈 맞춰줌
 body 에 모든 내용이 들어감.

```python
css = """

- 우리가 html 파일을 만들었다면 각각의 태그에 다르게 css 를 설정 할 것입니다. 
  이 때, 어느 요소에 스타일을 적용할지 알려주는 방식이 바로 css 선택자 입니다.

- 선택자가 어떻게 쓰이는가?

div {
    color: black;
    background-color: yellow;
}
div 태그들에게 css 를 주기 위해서 div 라는 선택자를 사용하였습니다. 이렇게 되면 html 문서에 
모든 div 태그들에는 { } 안에 적용해준 css 속성값들이 적용이 됩니다. div 처럼 태그의 이름 말고도 
선택자를 주는 방식이 굉장히 다양함

선택자의 종류

[ 선택자를 하나만 쓸 경우 ]

전체 선택자
전체 선택자는 * 을 사용하여 나타냅니다.

html 의 모든 태그에 대하여 적용합니다.

* {
    margin: 0 auto;
} 

태그 선택자
태그 선택자는 태그의 이름 을 사용하여 나타냅니다.
태그명이 선택자와 같은 태그들에 대하여 적용합니다.

p {
    color: red;
} 

클래스 선택자
클래스 선택자는 .클래스이름 을 사용하여 나타냅니다.
클래스가 클래스이름과 같은 태그들에 대하여 적용합니다.

.ClassName {
    width: 100px;
} 

ID 선택자
ID 선택자는 #아이디이름 을 사용하여 나타냅니다.
ID가 ID 이름과 같은 태그에 대해서 적용합니다.

#IdName {
    position: fixed;
} 

[ 선택자를 여러 개 쓸 경우] 

선택자를 여러 개 쓰는 경우를 복합 선택자라고 합니다. 복합 선택자에도 여러 가지 종류가 있어 
이에 대해서 알아보겠습니다.

하위 선택자
하위 선택자는 선택자 사이를 공백을 사용하여 나타냅니다.

앞 요소의 자손인 뒷 요소를 선택합니다.

section ul {
    text-shadow: none;
}
자식 선택자

하위 선택자는 선택자 사이를 > 를 사용하여 나타냅니다.

앞 요소의 자식 인 뒷 요소를 선택합니다.

section > ul {
    text-shadow: none;
}

자손과 자식의 차이는 무엇일까요?
자손은 자식을 포괄하는 의미입니다.
자손은 모든 하위 요소를 의미하고 자식은 바로 아래의 자식 요소에만 적용합니다.

인접 형제 선택자
인접 형제 선택자는 선택자 사이를 + 를 사용하여 나타냅니다.

앞 요소 바로 뒤의 뒷 요소를 선택합니다.

h1 + ul {
    color: red;
} 

일반 형제 선택자
일반 형제 선택자는 선택자 사이에 ~ 를 사용하여 나타냅니다.

앞 요소 뒤에 있는 모든 뒷 요소를 선택합니다.

h1 + ul {
    color: red;
}

형제라는 건 무엇일까요?

같은 부모를 가지는 요소들을 형제 라고 합니다.

속성 선택자
속성 선택자는 각 태그가 가지고 있는 그 속성에 접근하는 방식입니다. 생각 외로 활용도가 높으며 
크롤링를 할 경우에도 적절한 id 와 class 이름이 없을 경우, 이용합니다.

태그[속성]
속성이름 에 해당되는 속성을 가진 태그를 선택합니다.

a[href] {
    font-size: 10px;
}
태그[속성="변수"]
속성 의 속성값이 변수 인 태그를 선택합니다.

a[href="http://www.naver.com"] {
    color: black;
}

태그[속성~="변수"]
속성 의 속성값이 변수 를 포함하는 태그를 선택합니다.

a[href~="naver"] {
    color: black;
}

태그[속성^="변수"]
속성 의 속성값이 변수 로 시작하는 태그를 선택합니다.

a[href^="http"] {
    color: black;
}

태그[속성$="변수"]
속성 의 속성값이 변수 로 끝나는 요소를 선택합니다.

a[href$="com"]{
    color: black;
}

태그[속성*="변수"]
속성 의 속성값이 변수 를 포함하는 태그를 선택합니다.

a[href*="naver"] {
    color: black;
}

태그[속성~="변수"] 와 태그[속성*="변수"] 의 차이는 무엇일까요?
~ 은 포함여부를 단어를 기준으로 판단하고 * 는 문자열 기준으로 판단합니다.

다음과 같은 상황에서 navers 는 어떻게 될까요?
~ 은 naver와 navers 는 단어가 다르기 때문에 선택을 못하고
* 은 navers 는 naver 라는 문자열을 포함하기 때문에 선택을 하게 됩니다.

태그[속성|="변수"]
속성 의 속성값이 변수 이거나 변수 로 시작하는 태그를 선택합니다.

a[href|="http"]{
    color: black;
}
"""
print(f'CSS 선택자 \n {css}')
```

# 데이터 수집용 라이브러리


파이썬 데이터 수집용 라이브러리는 크게 두가지로 분류할 수 있습니다.

1. 파이썬 표준 라이브러리 모듈: 파이썬에서 기본적으로 제공되는 라이브러리 모듈로, urllib, xml, html.parser, json 등이 이에 해당합니다.

2. 외부 라이브러리 모듈: 다른 개발자가 만들어 공개한 라이브러리 모듈로, requests, BeautifulSoup, Scrapy, Selenium 등이 이에 해당합니다.


이러한 라이브러리 모듈들을 활용하면 데이터 수집과 가공을 빠르고 쉽게 수행할 수 있습니다.

- BeautifulSoup 모듈: HTML, XML 문서를 파싱하는 데 사용되는 라이브러리입니다.
- Scrapy 모듈: 웹 크롤링을 위해 만들어진 라이브러리로, 대규모 웹사이트에서 데이터 수집을 효과적으로 수행할 수 있습니다.
- Requests 모듈: HTTP 요청을 보내고 받는 데 사용되는 라이브러리입니다. ★REST API를 사용할 때 자주 사용됩니다.
- Selenium 모듈: 웹 브라우저를 제어하여 데이터를 수집하는 데 사용됩니다. 특히 동적인 웹사이트에서 데이터를 수집할 때 유용합니다.

```python
# 단순한 패턴을 읽을 때에는 복잡한 정규 표현식보다는 간단한 parse 모듈을 사용
# 파싱(parsing) : 문자열이나 파일 등의 입력을 받아서 그것을 더 작은 단위로 쪼개고 분석하여 
#그 구조를 익히는 방법


URL = 'https://section.blog.naver.com/Search/Post.naver?pageNo=1&rangeType=ALL&orderBy=sim&keyword=ai'
parse = urllib.parse.urlparse(URL)
print(parse,'\n')
print(parse[0],'\n')  #URL의 프로토콜 출력 ->https 출력
print(parse[1],'\n')  #URL의 네트워크 위치 ->section.blog.naver.com
print(parse[2],'\n')  # URL의 경로 -> /Search/Post.naver
print(parse[3],'\n')  #URL의 매개변수 ->pageNo=1&rangeType=ALL&orderBy=sim&keyword=ai
print(parse[4],'\n')  #URL의 쿼리 문자열 ->쿼리 문자열이 없으므로 빈 문자열('')이 출력
print(parse[5])       # URL의 프래그먼트 ->프래그먼트가 없으므로 빈 문자열('')이 출력
```

```python
# urllib.parse.urlsplit(): url을 5개로 분리하여 반환

URL = 'https://blog.naver.com/koreatech91;a=1?b=2#b'
parse1 = urllib.parse.urlparse(URL) # URL을 6개의 구성 요소로 분리하여 반환
parse2 = urllib.parse.urlsplit(URL) #URL을 5개의 구성 요소로 분리하여 반환
print(parse1,'\n')
print(parse2)
```

#### ♣ <font color=#6a51a3>두 코드의 차이점</font> ♣
##### <font color=#6baed6>urllib.parse.urlparse()</font>
- URL을 6개의 구성 요소로 분리하여 반환
     - scheme: URL의 스킴(protocol)부분을 반환합니다. (ex: "http", "https")
     - netloc: 네트워크 위치를 나타내는 부분(사용자 정보, 호스트 이름, 포트 번호를 포함)
     - path: 리소스 경로를 나타내는 부분으로, 계층 구조로 구성 (ex: "/path/to/resource")
     - params: 세미콜론(;)으로 구분된 파라미터를 나타내는 부분 (ex: ";param=value")
     - query: 물음표(?)로 구분된 쿼리 문자열을 나타내는 부분(ex: "key=value&key2=value2")
     - fragment: 샵(#)으로 구분된 프래그먼트 식별자를 나타내는 부분 (ex: "section")
     - URL에 포함된 파라미터(params)를 구분하여 반환하며, 반환값은 튜플
<br>  

###### <font color=#6baed6> urllib.parse.urlsplit()</font>
- URL을 5개의 구성 요소로 분리하여 반환
    - scheme: URL의 스킴(protocol)부분을 반환 (ex: "http", "https")
    - netloc: 네트워크 위치를 나타내는 부분(사용자 정보, 호스트 이름, 포트 번호를 포함)
    - path: 리소스 경로를 나타내는 부분으로, 계층 구조로 구성(ex: "/path/to/resource")
    - query: 물음표(?)로 구분된 쿼리 문자열을 나타내는 부분(ex: "key=value&key2=value2")
    - fragment: 샵(#)으로 구분된 프래그먼트 식별자를 나타내는 부분(ex: "section")
    - URL에 포함된 파라미터를 구분하지 않고, 반환값은 네임드 튜플(namedtuple)
<br>
<br>
###### urllib.parse.urlsplit() 함수는 urllib.parse.urlparse() 함수보다 반환하는 요소의 개수가 적지만, 반환값을 속성 이름으로 접근할 수 있다는 장점이 있습니다.

```python
# urllib.parse.urlunparse() 함수는 urllib.parse 모듈에 있는 함수 중 하나로, 
# URL의 구성 요소들을 입력받아 URL 문자열을 반환
```

```python
from urllib.parse import urlunparse

url_components = ('https', 'www.example.com', 'path/to/file', '', 'param1=value1&param2=value2', 'section')
url = urlunparse(url_components)
print(url)
```

```python
# urllib.parse.urlunsplit(): 분리된 url다시 합침(튜플로 반환)
url = 'https://blog.naver.com/koreatech91;a=1?b=2#b'
parse1 = urllib.parse.urlparse(url)
print(parse1[1])    #blog.naver.com

parse1 = list(parse1)   #튜플이어서 리스트로 변환해서 사용
parse1[1] = 'blog.daum.net'    #네이버->다음 주소로 수정

unparse1 = urllib.parse.urlunparse(parse1)
print(unparse1)

url_components = ('https', 'www.example.com', '/path/to/file', 'param1=value1&param2=value2', 'section')
urlunsplit1 = urllib.parse.urlunsplit(url_components)
print(urlunsplit1)
```

#### urljoin()

- 기본 url : 오리지널 url

```python
# 웹 스크래핑 - 네이버에서 파이썬 검색한 결과 한글로 추출 
from bs4 import BeautifulSoup
import requests
import re


URL = 'http://search.naver.com/search.naver?query="파이썬"'
req = requests.get(URL) # 검색어 파이썬을 포함한 url를 req에 저장.
# print(req.text)
soup =  BeautifulSoup(req.text, 'html.parser')
soup = str(soup) # bs로 파싱하고 str함수를 이용해 문자열로 변환
item = re.findall('[가-힣]+',soup) # 모든 한글을 찾는다
m = ' '.join(item) #' '안에 띄어쓰기 해줘야 출력물에 띄어쓰기 적용됨=join메소드로 띄어쓰기 합쳐줌
m
```

```python
#quote(), unquote(): 아스키코드가 아닌 퍼센트 인코딩으로 변환 
# 퍼센트 인코딩 (percent-encording)은 URL에 문자를 표현하는 문자 인코딩 방법
# url 에 한글이 섞이면 오류 발생



print(urllib.parse.quote('파이썬'))
#urllib.parse.quote() 메소드를 사용하여 '파이썬' 문자열을 URL 인코딩
#URL에서 한글 문자를 안전하게 전송하기 위해 필요한 작업


print(urllib.parse.unquote('%ED%8C%8C%EC%9D%B4%EC%8D%AC'))
#urllib.parse.unquote() 메소드를 사용하여 인코딩 된 문자열을 다시 디코딩하여 '파이썬' 문자열로 변환

URL = 'https://search.naver.com/search.naver?query=%ED%8C%8C%EC%9D%B4%EC%8D%AC'
# URL = 'https://search.naver.com/search.naver?query="파이썬"'

response = urllib.request.urlopen(URL)
# 디코딩 된 문자열을 URL의 쿼리 매개 변수로 사용하여 네이버 검색 엔진으로 GET 요청 보냄

byte_data = response.read() #URL의 HTML 응답을 읽어들여 바이트 데이터로 저장
text_data = byte_data.decode('utf-8') #바이트 데이터를 utf-8 인코딩을 사용하여 문자열로 디코딩
print(text_data) #'파이썬' 검색어를 사용하여 네이버 검색 엔진에서 반환된 검색 결과 페이지의 HTML 내용
```

```python
# 홈페이지를 로컬에 파일로 저장
# = 웹 페이지에서 HTML 데이터를 읽어들인 다음, 이를 'pc.html' 파일에 저장하는 기능

#웹 페이지에 대한 요청 객체를 생성
request = urllib.request.Request('http://www.naver.com') 

# 'urllib.request.urlopen()' 함수를 사용하여 요청 객체를 처리
#=>웹 페이지에서 반환된 바이트 데이터를 'data' 변수에 저장
data = urllib.request.urlopen(request).read()

# 'open()' 함수를 사용하여 'pc.html' 파일을 바이너리 쓰기 모드로 열기
f = open('pc.html','wb') 

#'f.write()' 메소드를 사용하여 'data' 변수에 저장된 바이트 데이터를 'pc.html' 파일에 쓰기
f.write(data)

#'f.close()' 메소드를 사용하여 파일 닫기 = 열었으면 꼭 닫아줘야 함
f.close()
```

```python
# 'with open()' 구문을 사용하여 'pc.html' 파일을 바이너리 모드로 오픈
#  = 파일을 열고 닫는 작업을 자동으로 처리하며, 파일을 닫지 않아도 됨
with open ('pc.html','rb') as f:
    data = f.read().decode('utf-8')
    print(data)
```

## requests 모듈
- 파이썬에서 HTTP 요청을 보내고 응답을 받는 기능을 제공하는 라이브러리
- 주요 메소드 
    - GET : 주어진 URL에서 데이터를 가져올 때 사용
    - POST : 주어진 URL에 데이터를 보낼 때 사용.
    - PUT : 주어진 URL에 데이터를 업로드 할 때 사용
    - DELETE : 주어진 URL에서 데이터를 삭제할 때 사용.

```python
# GET 메소드 사용 예제

import requests
response = requests.get('http://www.example.com/')
print(response.text)
```

```python
# POST 메소드 사용 예제

data =  {'username':'user', 'password':'pass'} #사용자 이름과 비번이 딕셔너리 형태로 저장. 
response = requests.post('https://www.example.com/login', data = data) 
print(response.status_code) 
#HTTP 상태 코드는 웹 서버가 요청을 받았을 때 처리 결과를 나타내는 세 자리 숫자로 구성
# 4xx = 사용자의 요청이 잘못됨
```

## urllib 하위 모듈

- urllib.request: 이 모듈은 URL을 열고 데이터를 읽어오는데 사용됩니다. <br>
    urlopen() 함수를 사용하여 URL에서 데이터를 가져올 수 있습니다. 이 모듈에는 HTTP,FTP, HTTPS 및 파일 URL을 처리하는 다양한 클래스와 함수가 포함되어 있습니다.
<br><br>
- urllib.parse : 이 모듈은 URL을 파싱하고 조작하는데 사용됩니다. urlparse() 함수를 사용하여 URL을 파싱하고, urlunparse() 함수를 사용하여 파싱된 URL을 다시 조립할 수 있습니다. 
<br><br>
- urllib.error : 이 모듈은 urllib.request 모듈에서 발생하는 예외를 처리하는데 사용됩니다. <br>
    RobotFileParser 클래스를 사용하여 robots.txt 파일을 파싱하고, can_fetch() 함수를 사용하여 주어진 로봇이 특정 URL에서 데이터를 가져올 수 있는지 여부를 확인할 수 있습니다.

```python
# 데이터를 가져오기 

import urllib.request

url = 'http://www.example.com'
response = urllib.request.urlopen(url)
html = response.read()
print(html)
```

```python
# Byte -> Text 로 변환

url = 'http://www.example.com'
response = urllib.request.urlopen(url)
byte_data = response.read(100)
print(byte_data,'\n')
text_data = byte_data.decode('utf-8')
print(text_data)
```

### BeautifulSoup 
- Python에서 HTML과 XML 문서를 파싱하여 원하는 정보를 추출하기 위한 라이브러리입니다.
- BeautifulSoup은 HTML 문서를 파싱하여 파이썬 객체로 변환합니다. 이 파이썬 객체를 다루면서 원하는 정보를 추출할 수 있습니다.
- BeautifulSoup을 사용하여 다음과 같은 작업을 수행할 수 있습니다.
    - HTML/XML 문서를 파싱하여 파이썬 객체로 변환하기
    - 파이썬 코드를 사용하여 문서 내부에서 데이터를 검색하고 추출하기
    - CSS 선택자와 XPath 표현식을 사용하여 문서 내부에서 원하는 요소를 선택하기
    - 선택한 요소의 속성과 값을 가져오기
    - 문서 내부에서 링크, 이미지 등의 다른 요소를 추출하기
    - 파싱된 데이터를 다시 HTML/XML 문서로 출력하기
- BeautifulSoup에서 제공하는 다양한 메서드

    - find() 메서드: HTML 문서에서 지정된 태그를 찾아내고 첫 번째로 발견된 태그를 반환합니다.

    - find_all() 메서드: HTML 문서에서 지정된 태그를 찾아내고, 모든 발견된 태그를 리스트로 반환합니다.

    - select() 메서드: CSS 선택자를 사용하여 HTML 문서에서 지정된 태그를 찾아내고, 모든 발견된 태그를 리스트로 반환합니다.

    - prettify() 메서드: 파싱된 HTML 문서를 보기 좋게 정리하여 반환합니다.

    - get() 메서드: HTML 태그의 속성(attribute) 값을 반환합니다.

    - text 속성: HTML 태그 안에 있는 텍스트를 추출합니다.

    - parent 속성: 현재 태그의 부모 태그를 반환합니다.

    - children 속성: 현재 태그의 자식 태그들을 리스트로 반환합니다.
    

BeautifulSoup 라이브러리를 사용하여 DOM 객체로 변환하고, 필요한 데이터를 추출
- 먼저 HTML 코드를 BeautifulSoup 객체로 변환한다. 이후, select 메서드를 사용하여 a 태그를 선택하고, href 속성을 사용하여 링크 URL을 추출한다. 이를 for 루프를 사용하여 출력할 수 있다.
- 따라서, 이와 같이 BeautifulSoup 라이브러리를 사용하여 DOM을 탐색하고 필요한 데이터를 추출할 수 있으며, 이를 이용하여 파이썬 웹 크롤러를 구현
- 위의 코드에서, HTML은 루트 요소인 $<html>$ 태그로 시작하며, 이 하위에 $<head>$ 태그와 $<body>$ 태그가 있다. $<html>$ 태그 안에는 문서 제목을 나타내는 $<title>$ 태그가 포함되어 있다.
- $<body>$ 태그는 HTML 문서에서 가장 중요한 부분으로, 페이지 내용을 담고 있는 부분이다. $<body>$ 태그 안에는 $<h1>$ 태그와 $<ul>$ 태그가 포함되어 있으며, $<h1>$ 태그는 페이지의 제목을 나타낸다. $<ul>$ 태그는 목록을 나타내며, 여기에는 3개의 $<li>$ 태그가 포함되어 있으며, 각각의 $<li>$ 태그 안에는 $<a>$ 태그가 포함되어 있다.

- 따라서, 위의 HTML 코드에서의 DOM 구조는 루트 요소인 $<html>$ 태그를 중심으로 계층 구조를 이루고 있으며, 이를 이용하여 자바스크립트나 파이썬 등의 스크립트 언어로 HTML 문서를 동적으로 조작하거나 데이터를 추출할 수 있다.

```python
# HTML 코드를 파싱하여 BeautifulSoup 객체로 만들기

from bs4 import BeautifulSoup

html = """
<!DOCTYPE html>
<html>
<head>
    <title> 웹 크롤링 예시 </title>
</head>
<body>
    <h1> 웹 크롤링 예시 </h1>
    <ul>
        <li><a href = "http://www.example.com"> Example 1 </a></li>
        <li><a href = "http://www.example.net"> Example 2 </a></li>
        <li><a href = "http://www.example.org"> Example 3 </a></li>
    </ul>
</body>
</html>

"""

soup = BeautifulSoup(html, 'html.parser')

# 링크 URL 추출하기

links = soup.select('a')
for link in links:
    print(link['href'])
```

```python
# 라이브러리 읽어 들이기

# next_sibling 메소드는 현재 선책된 요소의 다음 형제 요소를 선택하는 것. 여기서 두 번 사용 하는 이유는,
# 첫 번째 next_sibling 메소드로 선택되는 요소는 줄바꿈 문자와 같은 공백 문자열이기 때문임

from bs4 import BeautifulSoup

#분석하고 싶은 HTML

html = """
<html><body>
    <h1> 스크래핑이란? </h1>
    <p> 웹 페이지를 분석하는 것 </p>
    <p> 원하는 부분을 추출하는 것 </p>
</body></html>
"""

# HTML 분석하기
soup = BeautifulSoup(html, 'html.parser')

# 원하는 부분 추출하기 
h1 = soup.html.body.h1
p1 = soup.html.body.p
p2 = p1.next_sibling.next_sibling

# 요소의 글자 출력하기 

print(h1)
print("h1 = " + h1.string)
print("p = " + p1.string)
print("p = " + p2.string)
```

## requests와 BeautifulSoup을 사용하여 크롤링하는 예제

```python
import requests
from bs4 import BeautifulSoup

# 크롤링 할 웹 페이지의 URL을 지정합니다.
url = 'http://www.example.com/'

#requests 모듈을 사용하여 웹 페이지의 HTML 코드를 가져옵니다.
response = requests.get(url)
print(response)
html = response.text

# BeautifulSoup 객체를 생성합니다.
soup = BeautifulSoup(html,'html.parser')

#BeautifulSoup 객체의 메서드를 사용하여 필요한 정보를 추출합니다.
title = soup.find('title').text
paragraphs = soup.find_all('p')
links = [a['href'] for a in soup.select('a')]

# 추출한 정보를 출력합니다.

print('Title: ', title,'\n')

print('Paragraphs: ')
for p in paragraphs:
    print('-',p.text,'\n')
    
print('Links: ')
for link in links:
    print('-',link)
```

```python
# 정규 표현식

import requests
import re

url = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp"
data = requests.get(url).text
soup = BeautifulSoup(data,'html.parser')

wf = soup.find('wf').string
wf = re.sub('[^0-9가-힣]',' ',wf)
wf
```
