```python

np.random.seed(0)
ar=np.random.rand(1000,15)
columns =['id','gender','age','region','prod_22','prod_23','price_22','price_23 ','qty_22','qty_23','time_22 ','time_23','amount_22','amount_23','sales']
df=pd.DataFrame(ar,columns=columns)
df

#id : 1번 ~ 1000번 일련번호
df.id=np.random.randint(1,1001) 


#gender : 0,1 정수 난수 생성
df.gender=np.random.randint(0,2)  


#10~80사이 정수
ages=np.random.randint(10,81,size=len(df))
df.age=ages 


#region : 1 ~ 10 사이 정수 난수 생성
df.region=np.random.randint(1,11,size=len(df))


#prod_22 : 제품코드 11111 ~ 99999, 제품 수 5
df.prod_22=np.random.choice([11111,22222,33333,44444,55555],len(df))

#prod_23 : 제품코드 11111 ~ 99999, 제품 수 5
df.prod_23=np.random.choice([11111,22222,33333,44444,55555],len(df))

#price_22 : 1000 ~ 50000 사이 난수 생성
df.price_22=np.random.randint(1,51,1000)*1000

#price_23 : 1000 ~ 50000 사이 난수 생성
df.price_23=np.random.randint(1,51,1000)*1000

 #qty_22 : 1 ~ 100 사이 정수 난수 생성하여('19년 구매 수량)
df.qty_22=np.random.randint(1,101,size=len(df))

#qty_23 : 1 ~ 100 사이 정수 난수 생성하여('20년 구매 수량)
df.qty_23=np.random.randint(1,101,size=len(df)) 

#time_22 : 01 ~ 24 사이 정수 난수 생성('19년에 가장 자주 구매한 시간대)
df.time_22=np.random.randint(1,24,size=len(df)) 
#time_23 : 01 ~ 24 사이 정수 난수 생성('20년에 가장 자주 구매한 시간대)
df.time_23=np.random.randint(1,24,size=len(df)) 

#amount_22: price_22 * qty_22

df.amount_22=df.price_22 * df.qty_22
#amount_23: price_23 * qty_23
df.amount_23=df.price_23 * df.qty_23

#sales : 22년 대비 23년 구매금액이 증가면 1, 감소면 0
df.sales=np.where(((df.time_23-df.time_22)>0),1,0)
df.head()
```
