#### 판다스

https://pandas.pydata.org/docs/index.html

* 판다스는 구조화된 데이터 형식을 제공. 시리즈는 1차원 배열, 데이터프레임은 2차원 배열
* 라이브러리는 여러 종류의 class와 다양한 내장 함수로 구성. 시리즈와 데이터프레임은 대표적인 클래스 객체임
* 시리즈 인덱스는 데이터 값과 일대일 대응. 파이썬 딕셔너리와 비슷한 구조. pandas.Series(딕셔너리)

- Contents
 - Series(변환, 인덱스 구조, 원소 선택)
 - DataFrame(변환, 행인덱스/열이름 지정, 삭제,선택,추가, 변경, 전치, 인덱스 활용)

#### 데이터프레임
* 데이터프레임은 2차원 배열. R의 데이터프레임에서 유래.
* 데이터프레임의 열은 각각 시리즈 개체. 
* 시리즈를 열벡터라고 하면 데이터프레임은 여러개의 열벡터들이 같은 행 인덱스를
  기준으로 줄지어 결합된 2차원 벡터 또는 행렬.
* 선형대수학에서 열 벡터(m x 1 행렬)는 m 원소들의 단일 열 행렬
* 행 벡터(1 x m 행렬)은 m원소들의 단일 행 행렬.
* 리스트, 딕셔너리, ndarray 등 다양한 데이터로부터 생성
* 반대로 리스트, 딕셔너리, ndarray 등으로 변환될 수 있음

```python
R=데이터 분석하는 통계프로그램 ->파이썬에 비해 한계가 있음.
```

# 판다스

판다스(Pandas)는 데이터 분석과 관련된 작업에서 주로 사용되는 파이썬 라이브러리입니다. 판다스는 데이터를 다루기 쉽게 다양한 기능을 제공하며, 행과 열로 이루어진 2차원 데이터인 테이블(Table) 형태의 데이터를 다루는 데 특화되어 있습니다.

판다스를 사용하면 다음과 같은 작업을 수행할 수 있습니다.

- 데이터 불러오기 및 저장: CSV, Excel, SQL 등 다양한 형식의 데이터를 불러오거나 저장할 수 있습니다.

- 데이터 정제: 결측치, 이상치 등의 문제를 처리하거나 필요 없는 열을 제거하는 등의 데이터 정제 작업을 수행할 수 있습니다.

- 데이터 변환: 데이터의 형식을 변환하거나, 열을 추가하거나 삭제하는 등의 데이터 변환 작업을 수행할 수 있습니다.

- 데이터 분석: 데이터를 그룹화하거나 통계 정보를 계산하는 등의 데이터 분석 작업을 수행할 수 있습니다.

- 데이터 시각화: Matplotlib 라이브러리와 함께 사용하여 데이터를 그래프나 차트로 시각화할 수 있습니다.

따라서, 데이터를 다루는 작업에서 판다스는 매우 유용한 도구 중 하나입니다.
