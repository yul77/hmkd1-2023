## 미분

최적화를 하기 위해서는 입력값이 변했을 때 함수의 출력값이 어떻게 달라지는지를 정량적으로 알아야 하며 이러한 과정이 미분이다.

함수의 그래프상에 있는 한 점을 줌심으로 그래프를 무한히 확대해 보면 그래프의 모양이 직선에 가까워지는데 이때의 기울기를 미분이라고 한다. 이 직선은 같은 점을 기준으로 그린 그래프의 접선과 같다.

신경망에서 미분의 필요성 : https://gggggeun.tistory.com/120           https://gggggeun.tistory.com/121

미분 개념 : https://www.youtube.com/watch?v=kMZ3RJEaLE4

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rc('font', family='Malgun Gothic') #한글 폰트 설정
mpl.rcParams['axes.unicode_minus'] = False # 마이너스 단위 처리

def f(x):
    return x**3 - 3 * x**2 + x

x = np.linspace(-1,3,400)
y = f(x)

plt.figure(figsize=(10,6))

plt.plot(x,y)
plt.plot(0,0,'ro')
plt.plot(x,x,'r:') # : = 점선
plt.plot(1,-1,'go')

plt.plot(x, (3 * 1**2 - 6 * 1 + 1)*(x - 1)-1, 'g--')
# (3 * 1**2 - 6 * 1 + 1)*(x - 1)-1  은     y = 3(x-1)^2-1   을 나타냅
#(1,-1) 지점에서 시작하여 y=3(x-1)^2-1 그래프를 그린 

plt.xlim(-3.5, 5.5)
plt.ylim(-4, 2)
plt.xticks(np.arange(-3 , 6))
plt.yticks(np.arange(-4 , 2))

plt.annotate('', xy=(1, 0), xytext=(0, 0), arrowprops=dict(facecolor='gray'))
#plt.annotate 함수는 그래프 상에 주석(annotation)을 추가하는 함수
# 화살표를 그리기 위한 시작점을 (1,0)으로, 끝점을 (0,0)으로 설정
# 화살표의 속성을 지정하기 위해 arrowprops 매개변수를 사용
# arrowprops 매개변수에는 화살표의 색상을 'gray'로 설정
plt.annotate('', xy=(1, 1), xytext=(1, 0), arrowprops=dict(facecolor='gray'))

plt.annotate('', xy=(2, -1), xytext=(1, -1), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=(2, -3), xytext=(2, -1), arrowprops=dict(facecolor='gray'))

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("함수의 기울기")
plt.show()
```

#### arrowprops 
plt.annotate() 함수에서 사용되는 매개변수 중 하나로, 주석에 사용되는 화살표의 속성을 설정하는 데 사용됩니다. arrowprops 매개변수에는 딕셔너리(dictionary) 형태의 값이 전달되며, 다음과 같은 속성을 포함할 수 있습니다.

- facecolor: 화살표 내부의 색상
- edgecolor: 화살표 테두리의 색상
- alpha: 화살표의 투명도
- arrowstyle: 화살표 모양
- linewidth: 화살표 선의 굵기
- connectionstyle: 주석 텍스트와 화살표 사이의 연결선 모양 등

<br>예를 들어, arrowprops=dict(facecolor='gray', arrowstyle='->', linewidth=2)와 같이 
<br>arrowprops 매개변수에 딕셔너리를 전달하여 색상, 화살표 모양, 선 굵기 등의 속성을 지정할 수 있습니다.

```python
# y = f(x) = x^3 - x 시각화 하는 코드

def f(x):
    return (x**3-x)
plt.figure(figsize=(15, 5)) #너비15, 높이 5
delta = [2.0, 0.2, 0.01]#elta 값이 클수록 함수 그래프가 더 거칠게 표현되며, 
                          #작을수록 함수를 더 상세하고 정확하게 볼 수 있다.

x = np.linspace(0.5-delta[0], 0.5+delta[0], 200) # 0.5-delta[0]에서 0.5+delta[0]까지 200개의 값으로 생성
y = f(x)

plt.subplot(131) #하나의 그림을 여러 개의 서브플롯(subplot)으로 나눌 때 사용하는 함수
                 #131은 현재 그림을 1행 3열로 나눈 후, 
                 #첫 번째 서브플롯에 해당하는 영역에 그래프를 그리겠다는 뜻

plt.ylim(-3.0/8.0-delta[0], -3.0/8.0+delta[0]) #plt. lim ()은 축의 범위설정 함수
    # 첫번째 인자는 y축 최소값 ,  두번째는 y축의 최대값
    
plt.xlim(0.5-delta[0], 0.5+delta[0])
        # x축의 최소값 , x 축의 최대값
plt.plot(x, y, 'b-', lw=1, c='k')  

plt.scatter([0.5], [-3.0/8.0])
plt.xlabel('x',fontsize=14)
plt.ylabel('y',fontsize=14)
plt.grid()
plt.title('delta = %.4f' % delta[0], fontsize=14)


x = np.linspace(0.5-delta[1], 0.5+delta[1], 200)
y = f(x)

plt.subplot(132)

plt.ylim(-3.0/8.0-delta[1], -3.0/8.0+delta[1])
plt.xlim(0.5-delta[1], 0.5+delta[1])
plt.plot(x, y, 'b-', lw=1, c='k')
plt.scatter([0.5], [-3.0/8.0])
plt.xlabel('x',fontsize=14)
plt.ylabel('y',fontsize=14)
plt.grid()
plt.title('delta = %.4f' % delta[1], fontsize=14)

x = np.linspace(0.5-delta[2], 0.5+delta[2], 200)
y = f(x)

plt.subplot(133)

plt.ylim(-3.0/8.0-delta[2], -3.0/8.0+delta[2])
plt.xlim(0.5-delta[2], 0.5+delta[2])
plt.plot(x, y, 'b-', lw=1, c='k')
plt.scatter([0.5], [-3.0/8.0])
plt.xlabel('x',fontsize=14)
plt.ylabel('y',fontsize=14)
plt.grid()
plt.title('delta = %.4f' % delta[2], fontsize=14)

plt.show()
```

### 접선의 방정식

- 함수 f(x)에서 x의 값을 dx만큼 변화시켰을 때 f(x)의 변화량 (f(x+dx)-f(x))는 f'(x)dx와 같다.
- 미분 : y=f(x)라는 그래프상의 점(a,f(a))가 있을 때 그 점에서 그은 접선의 기울기는 f'(a)이며 a에서의 [순간 변화율]을 의미 (=미분의 핵심!)
- 적분 : 도함수 f'(x)의 하방 면적은 원 함수 f(x)의 y의 값이다. ex) f(x) = $x^2$

```python
def f(x):
    return (x**2)
def g(x):
    return (2*x - 1)

x = np.linspace(-3, 3, 100)
fig = plt.figure(figsize = (5,5))

plt.scatter([1],[1],c='k') # 점.

plt.plot(x,f(x),'b-',lw=1,c='k') #x에 대해서 f(x)함수를 정해줌
plt.plot(x,g(x),'b-',lw=1,c='b') #기울기

plt.plot([x.min(),x.max()],[0,0],lw=2, c='k')
plt.plot([0,0],[g(x).min(), f(x).max()],lw=2, c='k')
plt.ylim(-1,3)
plt.xlim(-2,2)
plt.grid(lw=1)
plt.show()
```

```python
# derivative: 주어진 함수의 도함수(derivative) 값을 계산해주는 함수

from scipy.misc import derivative

print(derivative(f,1,dx=1e-6))
print(derivative(f,2,dx=1e-6))
```

### 편미분
- 다변수 함수에서 한 변수를 제외한 나머지 변수들의 값을 고정하고, 해당 변수의 변화에 따른 함수값의 변화량을 나타내는 개념

```python
# Juypter 노트북에서 수학식의 LaTeX 표현을 위해 필요함
# 심볼릭 변수는 수식과 연산에 사용되며, 수식을 미분하거나 적분하는 등의 연산을 수행할 때 사용
import sympy

sympy.init_printing(use_latex = 'mathjax') #use_latex 인자를 'mathjax'로 설정 = LaTeX 수식 표현을 사용
# sympy.init_printing 함수 :수식을 더 보기 좋게 출력하기 위한 설정
#

x = sympy.symbols('x')
x
```

```python
x,y = sympy. symbols('x y')
f = x ** 2 + 4 * x * y + 4 * y ** 2
f
```
