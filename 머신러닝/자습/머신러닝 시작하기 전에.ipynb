## 넘파이 복습

```python
import numpy as np

array1 = np.array([1,2,3])
print('array1 type: ',type(array1))
print('array1 array 형태:', array1.shape) # 1차원 array로 3개의 데이터를 가지고 있음.
                                          # 1차원임을 (3,)로 표현함

array2 = np.array([[1,2,3],
                   [2,3,4]])
print('array2 type:',type(array2))
print('array2 array 형태:', array2.shape) # 2차원 array로 2개의 로우와 3개의 컬럼 보유.

array3 = np.array([[1,2,3]])
print('array3 type: ',type(array3))
print('array3 array 형태:', array3.shape) # 2차원 array로 1개의 로우와 3개의 칼럼 보유
                                          # 2차원임을 (1,3)으로 표현함.
```

```python
# array 별 몇차원인지 출력

print('array1: {:0}차원, array2: {:1}차원, array3: {:2}차원'.format(array1.ndim, array2.ndim, array3.ndim))
```

```python
# ndarray 내의 데이터 타입은 dtype속성으로 확인 가능

list1 = [1,2,3] # 리스트 자료형
print(type(list1))

array1 = np.array(list1)  #ndarray로 변경
print(type(array1))

print(array1, array1.dtype)
```

```python
# 서로다른 데이터 타입을 가질 수 있는 리스트와 달리
# ndarray내의 데이터 타입은 그 연산 특성상 같은 데이터 타입만 가능


# 문자와 숫자가 섞여 있을 경우.
list2 = [1,2,'test']
array2 = np.array(list2)
print(array2,array2.dtype)
#데이터 타입이 더 큰 쪽으로 변환이 되어 숫자형이 모두 유니코드문자열로 변경됨

list3 = [1,2,3.0]
array3 = np.array(list3)
print(array3, array3.dtype)
# int 1,2가 모두 float64형으로 변경됨
```

```python
array_int = np.array([1,2,3])
array_float = array_int.astype('float64')
print(array_float, array_float.dtype)
array_int1 = array_float.astype('int32')
print(array_int1, array_int1.dtype)

array_float1 = np.array([1.1,2.1,3.1])
array_int2 =  array_float1.astype('int32')
print(array_int2, array_int2.dtype)
```

```python
# np.argsort()
# 원본 행렬이 정렬되었을 때 [기존 행렬의 원소에 대한 인덱스를 필요로 할 때 사용.]
# = 정렬 행렬의 원본 행렬 인덱스를 ndarray 형으로 반환

org_array = np.array([3,1,9,5])
sort_indices = np.argsort(org_array)
print(type(sort_indices))
print('행렬 정렬 시 원본 행렬의 인덱스: ',sort_indices)
```

```python
org_array = np.array([3,1,9,5])
sort_indices_desc = np.argsort(org_array)[::-1]
print('행렬 내림차순 정렬 시 행렬의 인덱스 :',sort_indices_desc)
```

```python
# np.argsort()를 학생데이터에 활용하는 예시

name_array = np.array(['John', 'Mike', 'Sarah', 'Kate', 'Samuel'])
score_array= np.array([78, 95, 84, 98, 88])

sort_indices_asc = np.argsort(score_array)
print('성적 오름차순 정렬 시 score_array의 인덱스: ',sort_indices_asc)
print('성적 오름차순으로 name_array의 이름 출력:', name_array[sort_indices_asc])
```

```python
# 팬시 인덱스
# 해당 위치의 인덱스에 해당하는 ndarray를 반환하는 인덱싱방법
import numpy as np
array1d = np.arange(start = 1, stop = 10)
array2d = array1d.reshape(3,3)

array3 = array2d[[0,1],2] # array2d에서 0번째와 1번째 행의 2번째 열 요소를 선택
print('array2d[[0.1],2]=>', array3.tolist()) # array3.tolist()= array3를 파이썬리스트로 출력

# 1  2   [3]
# 4  5   [6]
# 7  8    9

array4 = array2d[[0,1],0:2]
print('array2d[[0.1],2]=>', array4.tolist())

# [1  2 ]  3
# [4  5 ]  6
#  7  8    9

array5 = array2d[[0,1]]
print('array2d[[0,1]]=>',array5.tolist())
# [1  2  3]
# [4  5  6]
#  7  8  9

# tolist()= 배열 형태의 객체를 파이썬 리스트로 변환하는 메소드
```

```python
# 불린 인덱싱
# 조건 필터링과 검색을 동시에 할수 있음.
# 인덱스를 지정하는 [] 안에 조건문을 그대로 기재하면 됨.

array1d = np.arange(start=1, stop=10)
# [ ]안에 array1d > 5 Boolean indexing을 적용
array3 = array1d[array1d > 5]
print('array1d > 5 불린 인덱싱 결과 값:', array3)
```

```python
array1d > 5
# 5보다 큰 값일 경우 True로 표기되고 있다.
# false가 있는 0~4 인덱스는 무시. [5,6,7,8] 인덱스의 실제값인 [6,7,8,9]를 반환하게 된다.
```

```python
# 위와 동일한 불린 ndarray를 만들고 이를 array1d[]내에 인덱스로 입력하면 동일한 데이터 세트가
# 변환됨을 알 수 있다.

boolean_indexes = np.array([False, False, False, False, False,  True,  True,  True,  True])
array3 = array1d[boolean_indexes]
print('불린 인덱스로 필터링 결과:', array3)
```

```python
# 즉, 다음과 같이 직접 인덱스 집합을 만들어 대입한 것과 동일 -- 33p참고

indexes = np.array([5,6,7,8])
array4 = array1d[indexes]
print('일반 인덱스로 필터링 결과:', array4) #5,6,7,8의 인덱스값을 출력
```

## np.sort()와 ndarray.sort(), argsort()의 차이
- np.sort()는 원 행렬은 그대로 유지한 채 원 행렬의 정렬된 행렬을 반환
- ndarray.sort()는 원 행렬 자체를 정렬한 형태로 변환하며 반환값은 None이다
- np.argsort() 는  정렬시 원본데이터의 인덱스 넘버출력

```python
# np.sort()로 정렬
org_array = np.array([3,1,9,5])
print('원본행렬:', org_array,'\n')

sort_array1 = np.sort(org_array)
print('np.sort() 호출 후 반환된 정렬 행렬:',sort_array1)
print('np.sort() 호출 후 원본 행렬:',org_array)
```

```python
# ndarray.sort() 로 정렬
sort_array2 = org_array.sort()
print('org_array.sort() 호출 후 반환된 행렬:', sort_array2)
#원본 배열을 인자로 받아 해당 배열을 오름차순으로 정렬합니다.
#정렬된 결과는 원본 배열 자체에 반영됩니다. 반환 값으로는 정렬된 결과가 아닌 None이 반환

print('org_array.sort() 호출 후 원본 행렬:',org_array)

# np.sort()는 원본 행렬을 변경하지 않고 정렬된 형태로 반환
# ndarray.sort()는 원본 행렬 자체를 정렬한 값으로 변환
```

```python
# 내림차순으로 정렬
sort_array1_desc = np.sort(org_array)[::-1]
print('내림차순으로 정렬:',sort_array1_desc)
```

```python
# 정렬된 행렬의 인덱스 반환 -35p
# 원본이 정렬되었을 떄 기존 원본 행렬의 원소에 대한 인덱스가 필요할때, np.argsort() 사용

org_array = np.array([3,1,9,5])
sort_indices = np.argsort(org_array)
print(type(sort_indices))
print('행렬 정렬 시 원본 행렬의 인덱스:', sort_indices)
sort_indices_desc = np.argsort(org_array)[::-1]
print('행렬 내림차순 정렬 시 원본 행렬의 인덱스:', sort_indices_desc)
```

```python
# 시험 성적순으로 학생 이름 출력

import numpy as np

name_array = np.array(['유리','슬기','인혁','수현','용훈'])
score_array = np.array([78,95,84,98,88])

sort_indices_asc = np.argsort(score_array)
print('성적 오름차순 정렬 시 score_array의 인덱스:', sort_indices_asc)
print('성적 오름차순으로 점수 출력:',np.sort(score_array))
print('성적 오름차순으로 name_array의 이름 출력:', name_array[sort_indices_asc])
```

## 선형대수 연산 - 행렬 내적과 전치 행렬 구하기

- 행렬 내적(=행렬의 곱)
: 두 행렬 A와 B의 내적은 np.dot()을 이용해 계산가능.



```python
A = np.array([[1,2,3],
              [4,5,6]])

B = np.array([[7,8],
              [9,10],
              [11,12]])

dot_product =  np.dot(A,B)
print('행렬 내적 결과 : \n',dot_product)

# 행렬의 1행 1열의 값인 58은 A행렬의 1행과 B행렬의 1열의 원소들을 차례로 곱한 값을 더한 것.
# 1 x 7 + 2 x 9 + 3 x 11 = 58
```

### 전치행렬
- 원 행렬에서 행과 열 위치를 교환한 행렬을 전치행렬 이라고 한다.

```python
A = np.array([[1,2],
              [3,4]])
transpose_mat = np.transpose(A)
print('A의 전치 행렬: \n', transpose_mat)
```

### DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환
- 넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기

```python
# 1차원 형태


col_name1 = ['col1']
list1 = [1,2,3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

# 리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1, columns = col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

#넘파이 ndarray를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n',df_array1)
```

```python
# 2차원 형태의 2행 3열

# 3개의 칼럼명이 필요함.
col_name2=['col1', 'col2', 'col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환.
list2 = [[1, 2, 3],
         [11, 12, 13]]
array2 = np.array(list2)
print('array2 shape:', array2.shape )
df_list2 = pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
df_array2 = pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```
